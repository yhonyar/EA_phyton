Primera forma

Se define las funciones que convertiran a los datos extraidos de los diferentes formatos (.csv, .json, .xml) en un ``DataFrame``

```python
def extraer_de_csv(archivo_procesar):
    df = pd.read_csv(archivo_procesar)
    
    filename = archivo_procesar.replace('./data_H5\\', '')
    filename2 = filename.replace('.csv', '')
    df['origen'] = filename2
    df['tipo'] = 'csv'
    
    return df
```

```python
def extraer_de_json(archivo_procesar):
    df = pd.read_json(archivo_procesar, lines = True)
    
    filename = archivo_procesar.replace('./data_H5\\', '')
    filename2 = filename.replace('.json', '')
    df['origen'] = filename2
    df['tipo'] = 'json'
    
    return df
```

```python
import xml.etree.ElementTree as et
import glob
```

```python
def extraer_de_xml(archivo_procesar):
    xtree = et.parse(archivo_procesar)
    xroot = xtree.getroot()
    df = pd.DataFrame(columns = ['nombre','ventas1','ventas'])
    for persona in xroot:
        nombre = persona.find('nombre').text
        ventas1 = float(persona.find('ventas1').text)
        ventas = float(persona.find('ventas').text)
        df = df.append({'nombre':nombre,'ventas1':ventas1,'ventas':ventas}, ignore_index = True)
        
    filename = archivo_procesar.replace('./data_H5\\', '')
    filename2 = filename.replace('.xml', '')
    df['origen'] = filename2
    df['tipo'] = 'xml'
    
    return df
```

```python
def extraer():
    extraer_dato = pd.DataFrame(columns = ['nombre','ventas1','ventas'])
    
    for a_csv in glob.glob('./data_H5/*.csv'):
        extraer_dato = extraer_dato.append(extraer_de_csv(a_csv), ignore_index = True)
        
    for a_json in glob.glob('./data_H5/*.json'):
        extraer_dato = extraer_dato.append(extraer_de_json(a_json), ignore_index = True)
        
    for a_xml in glob.glob('./data_H5/*.xml'):
        extraer_dato = extraer_dato.append(extraer_de_xml(a_xml), ignore_index = True)
        
    return extraer_dato
```

```python
extraer()
```

```python
from pathlib import Path
```

```python
def extraer_csv1(archivo_procesar):
    df = pd.read_csv(archivo_procesar)
    df['origen'] =Path(archivo_procesar).stem
    return df
```

```python
def extraer_json1(archivo_procesar):
    df = pd.read_json(archivo_procesar, lines = True)
    df['origen'] =Path(archivo_procesar).stem
    return df
```

```python
def extraer_xml1(archivo_procesar):
    xtree = et.parse(archivo_procesar)
    xroot = xtree.getroot()
    df = pd.DataFrame(columns = ['nombre','ventas1','ventas']) 
    for persona in xroot:
        nombre = persona.find('nombre').text
        ventas1 = float(persona.find('ventas1').text)
        ventas = float(persona.find('ventas').text)
        df = df.append ({'nombre':nombre,'ventas1':ventas1,'ventas':ventas},ignore_index = True)
        df['origen'] =Path(archivo_procesar).stem
        return df
```

```python
def extraer1():
    extraer_dato = pd.DataFrame(columns = ['nombre','ventas1','ventas'])
    
    for a_csv in glob.glob('./data_H5/*.csv'):
        p = extraer_csv1(a_csv)
        p['tipo'] = "csv"
        extraer_dato = extraer_dato.append(p, ignore_index = True) 
              
    for a_json in glob.glob('./data_H5/*.json'):
        p = extraer_json1(a_json)
        p['tipo'] = "json"
        extraer_dato = extraer_dato.append(p, ignore_index = True)
        
    for a_xml in glob.glob('./data_H5/*.xml'):
        p = extraer_xml1(a_xml)
        p['tipo'] = "xml"
        extraer_dato = extraer_dato.append(p, ignore_index = True)
    return extraer_dato
```
